```python
import pandas as pd
from gensim.models import Word2Vec
```

```python
model = Word2Vec.load('../ko/ko.bin')
```

```python
food_df = pd.read_csv("../data/food_keyword1.csv")
food_df.head(5)
```

```python
df_col = food_df.columns
for idx in food_df.index:
    print("------" + food_df[df_col[0]][idx] + "------")
    for keyword in food_df[df_col[1]][idx].split(" "):
        try:
            print(keyword , " : " , model.wv.most_similar(keyword,topn=4))
        except:
            print(f'{keyword} : None')
```

```python
def check_word(word):
    print(model.wv.most_similar(word,topn=20))
```

```python
check_word('비타민')
```

```python
check_word('항염')
```
