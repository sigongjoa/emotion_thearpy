```python
import pandas as pd
from gensim.models import Word2Vec
```

```python
model = Word2Vec.load('../ko/ko.bin')
```

# 차원 유사어에 대해서 w2v 존재 여부 확인

```python
energy = ['정열', '활기', '기력', '생기', '활발']
recovery = ['낫다', '치료', '쉼' ,'극복' , '재활']
cycle = ['돌다', '유지', '조절' , '공급' ,'생명']
pury = ['진정' , '맑다' , '없애다' , '깨끗하다', '해소'] 
all_dims = energy + recovery + cycle + pury       
```

```python
for dim_word in all_dims:
    try:
        print(model.wv.most_similar(dim_word,topn=2))
    except:
        print(f'{dim_word} is None')
```

```python
def check_word(word):
    print(model.wv.most_similar(word,topn=20))
```

```python
check_word('낫')
```

```python
check_word('쉼')
```

```python
check_word('맑')
```

```python
check_word('없애')
```

```python
check_word('깨끗하')
```

# 활동 키워드 에 대해서 w2v 존재 여부 확인

```python
dir = "../data/action_keyword_data.csv"
data = pd.read_csv(dir)
data
```

```python
for idx in data.index:
    act_name = data['action'][idx]
    act_effs = data['효능'][idx].split(" ")
    print(f'----{act_name}-----')
    for act_eff in act_effs: 
        try:
            print(act_eff , model.wv.most_similar(act_eff,topn=2))
        except:
            print(f'{act_eff} is None')
```
